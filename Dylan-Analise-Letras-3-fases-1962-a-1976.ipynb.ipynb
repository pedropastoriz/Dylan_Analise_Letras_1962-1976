```python
import os
import requests
from bs4 import BeautifulSoup

# Função para raspar e salvar letra
def baixar_letra(url, nome_musica, pasta_destino):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Pegando o conteúdo da letra
    div_letra = soup.find('div', class_='article-content lyrics')
    if div_letra:
        letra = div_letra.get_text(separator='\n').strip()

        # Criando diretório se não existir
        os.makedirs(pasta_destino, exist_ok=True)

        # Nome do arquivo
        caminho_arquivo = os.path.join(pasta_destino, f"{nome_musica}.txt")

        # Salvando
        with open(caminho_arquivo, 'w', encoding='utf-8') as f:
            f.write(letra)
        print(f"Letra de '{nome_musica}' salva em {caminho_arquivo}")
    else:
        print(f"Letra não encontrada em {url}")
```
